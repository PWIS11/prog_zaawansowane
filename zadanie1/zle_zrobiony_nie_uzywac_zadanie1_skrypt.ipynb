## actual_generation

```python
na_symbols = ["--", "-", "NA", "N/A", ""]
actual_generation = pd.read_csv("Actual_generation.csv", sep=';',  na_values=na_symbols)
```

```python
print(actual_generation.isnull().sum())
print(actual_generation.dtypes)
#print(actual_generation.count())
```

```python
cols_to_fix_gener = [
    'Biomass [MWh] Calculated resolutions', 
    'Hydropower [MWh] Calculated resolutions',
    'Wind offshore [MWh] Calculated resolutions',
    'Wind onshore [MWh] Calculated resolutions',
    'Photovoltaics [MWh] Calculated resolutions',
    'Other renewable [MWh] Calculated resolutions',
    'Nuclear [MWh] Calculated resolutions',
    'Lignite [MWh] Calculated resolutions',
    'Hard coal [MWh] Calculated resolutions',
    'Fossil gas [MWh] Calculated resolutions',
    'Hydro pumped storage [MWh] Calculated resolutions',
    'Other conventional [MWh] Calculated resolutions'
]

for col in cols_to_fix_gener:
    actual_generation[col] = actual_generation[col].str.replace(',', '').astype(float)

print(actual_generation.dtypes)
```

Typy danych w naszym zbiorze: object zostały zamienione na float. Natomiast braki danych zostały nie ruszone oprócz kolumny "Nuclear [\MWh] Calculated resolutions", w której zaobserwowano 24 brakujące obserwacje. Po podejrzeniu w/w kolumny widzimy, ze braki w danych występują kolejno po sobie, zatem zakładamy, ze występował wtedy postój elektrowni i uzupełniamy ją zerami

```python
actual_generation['Nuclear [MWh] Calculated resolutions'] = actual_generation['Nuclear [MWh] Calculated resolutions'].fillna(0)

print(f"Braki w Nuclear po naprawie: {actual_generation['Nuclear [MWh] Calculated resolutions'].isnull().sum()}")
print(f"Ostatnie wartości:\n{actual_generation['Nuclear [MWh] Calculated resolutions'].tail()}")
```

## Balancing energy

```python
na_symbols = ["--", "-", "NA", "N/A", ""]
balancing =  pd.read_csv("Balancing_energy.csv", sep=';',  na_values=na_symbols)
```

```python
print(balancing.isnull().sum())
```

```python
print(balancing.dtypes)
```

```python
cols_to_fix_balancing = [
    'Volume (+) [MWh] Calculated resolutions', 
    'Volume (-) [MWh] Calculated resolutions',
    'Net income [€] Original resolutions'
]

for col in cols_to_fix_balancing:
    balancing[col] = balancing[col].str.replace(',', '').astype(float)

print(balancing.dtypes)
```

```python
cols_to_fix_balancing_withoutvolume = [
    'Net income [€] Original resolutions'
]
balancing[cols_to_fix_balancing_withoutvolume] = balancing[cols_to_fix_balancing_withoutvolume].fillna(balancing[cols_to_fix_balancing_withoutvolume].median())

print("Liczba braków po uzupełnieniu:")
print(balancing[cols_to_fix_balancing_withoutvolume].isnull().sum())
```
